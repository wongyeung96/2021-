### 1. 파일 불러오기

```python
import pandas as pd
```

```python
_0_5 = pd.read_csv("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/4. 결론 1 최종 버퍼별 순위 csv+top20 geojson/500m 순위 2.csv",index_col = 'Unnamed: 0',encoding='cp949')
_1 = pd.read_csv("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/4. 결론 1 최종 버퍼별 순위 csv+top20 geojson/1km 순위 2.csv",index_col = 'Unnamed: 0',encoding='cp949')
_1_5 = pd.read_csv("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/4. 결론 1 최종 버퍼별 순위 csv+top20 geojson/1_5km 순위 2.csv",index_col = 'Unnamed: 0',encoding='cp949')
_2 = pd.read_csv("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/4. 결론 1 최종 버퍼별 순위 csv+top20 geojson/2km 순위 2.csv",index_col = 'Unnamed: 0',encoding='cp949')
```

### 2. 데이터 concat

```python
df = pd.concat([_0_5,_1], ignore_index=True)
df = pd.concat([df,_1_5], ignore_index=True)
df = pd.concat([df,_2], ignore_index=True)
df
```

### 3. 점수 합하고 정렬

```python
score = df.groupby('address').sum()
```

```python
score.sort_values(by='score',ascending=False,inplace=True)
```

```python
score.reset_index()
```

### 4. geojson 파일로 저장하기

```python
import geopandas as gpd
lpg=gpd.read_file("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터(정제데이터)/1.QGIS에 사용할 데이터(gid 가능데이터)/0. new lpglocation.geojson")
lpg.rename(columns ={"소재지":"address"},inplace = True)
score= score.merge(lpg,on='address',how='inner')
```

```python
score
```

```python
score_geo = gpd.GeoDataFrame(score)
score_geo.to_file('C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/4. 결론 2 버퍼별 점수 합산으로 최적 입지 선정 geojson/LPG 버퍼 총 합산 점수 2.geojson', driver='GeoJSON') 
```
